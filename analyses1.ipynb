```python
# DÃ©finition du chemin vers les donnÃ©es
data_path = Path("data")

# On commence par le fichier des chantiers Ã  Paris
chantiers_file = data_path / "comptage-velo-donnees-compteurs.csv"

# VÃ©rification que le fichier existe
if chantiers_file.exists():
    print(f"âœ… Fichier trouvÃ© : {chantiers_file}")
else:
    print(f"âŒ Fichier non trouvÃ© : {chantiers_file}")
    # Liste des fichiers disponibles
    print("\nğŸ“ Fichiers disponibles dans data/ :")
    for f in data_path.glob("*.csv"):
        print(f"   - {f.name}")
```

```python
# Fonction pour dÃ©tecter automatiquement le sÃ©parateur
def detect_separator(file_path, sample_lines=5):
    """DÃ©tecte le sÃ©parateur utilisÃ© dans un fichier CSV"""
    with open(file_path, 'r', encoding='utf-8') as f:
        first_line = f.readline()
        # Compte les occurrences de chaque sÃ©parateur possible
        separators = {';': first_line.count(';'), ',': first_line.count(',')}
        # Retourne le sÃ©parateur le plus frÃ©quent
        return max(separators, key=separators.get)

# Chargement du fichier CSV avec dÃ©tection automatique du sÃ©parateur
try:
    # DÃ©tection du sÃ©parateur
    sep = detect_separator(chantiers_file)
    print(f"ğŸ” SÃ©parateur dÃ©tectÃ© : '{sep}'")
    
    # Chargement avec gestion des erreurs
    df_chantiers = pd.read_csv(
        chantiers_file,
        sep=sep,
        encoding='utf-8',
        low_memory=False,
        on_bad_lines='skip'  # Ignore les lignes problÃ©matiques
    )
    
    print(f"âœ… Fichier chargÃ© avec succÃ¨s")
    print(f"ğŸ“Š Dimensions : {df_chantiers.shape[0]} lignes Ã— {df_chantiers.shape[1]} colonnes")
    
except Exception as e:
    print(f"âŒ Erreur lors du chargement : {e}")
    print("\nğŸ”„ Tentative avec diffÃ©rents sÃ©parateurs...")
    
    # Essai avec point-virgule
    try:
        df_chantiers = pd.read_csv(chantiers_file, sep=';', encoding='utf-8', low_memory=False, on_bad_lines='skip')
        print("âœ… ChargÃ© avec succÃ¨s en utilisant le sÃ©parateur ';'")
        print(f"ğŸ“Š Dimensions : {df_chantiers.shape[0]} lignes Ã— {df_chantiers.shape[1]} colonnes")
    except:
        # Essai avec virgule
        try:
            df_chantiers = pd.read_csv(chantiers_file, sep=',', encoding='utf-8', low_memory=False, on_bad_lines='skip')
            print("âœ… ChargÃ© avec succÃ¨s en utilisant le sÃ©parateur ','")
            print(f"ğŸ“Š Dimensions : {df_chantiers.shape[0]} lignes Ã— {df_chantiers.shape[1]} colonnes")
        except Exception as e2:
            print(f"âŒ Ã‰chec : {e2}")
```

```python
import pandas as pd
from pathlib import Path
import os

# Configuration des chemins
data_path = Path("data")
output_path = Path("mydata")
output_path.mkdir(exist_ok=True)

input_file = data_path / "comptage-velo-donnees-compteurs.csv"
output_file = output_path / "comptage-velo-donnees-compteurs-cleaned.csv"

# Colonnes Ã  conserver (essentielles et utiles)
colonnes_a_conserver = [
    'Identifiant du compteur',
    'Nom du compteur',
    'Identifiant du site de comptage',
    'Nom du site de comptage',
    'Comptage horaire',
    'Date et heure de comptage',
    'Date d\'installation du site de comptage',
    'CoordonnÃ©es gÃ©ographiques',
    'Identifiant technique compteur',
    'mois_annee_comptage'  # On garde pour faciliter les agrÃ©gations
]

print(f"ğŸ“‚ Lecture du fichier : {input_file}")
print(f"ğŸ“Š Taille du fichier : {input_file.stat().st_size / (1024**3):.2f} GB")

# Lecture par chunks pour gÃ©rer le gros fichier
chunk_size = 50000  # Lire 50k lignes Ã  la fois
chunks = []

print("\nğŸ”„ Traitement du fichier par chunks...")
for i, chunk in enumerate(pd.read_csv(input_file, sep=';', chunksize=chunk_size, low_memory=False)):
    # Nettoyer les noms de colonnes (supprimer BOM si prÃ©sent)
    chunk.columns = chunk.columns.str.replace('\ufeff', '')
    
    # SÃ©lectionner uniquement les colonnes Ã  conserver (qui existent)
    colonnes_existantes = [col for col in colonnes_a_conserver if col in chunk.columns]
    chunk_cleaned = chunk[colonnes_existantes]
    
    chunks.append(chunk_cleaned)
    
    if (i + 1) % 10 == 0:
        print(f"   âœ“ TraitÃ© {len(chunks) * chunk_size:,} lignes...")

print(f"\nğŸ“ ConcatÃ©nation des {len(chunks)} chunks...")
df_cleaned = pd.concat(chunks, ignore_index=True)

print(f"âœ… Fichier nettoyÃ© : {df_cleaned.shape[0]:,} lignes Ã— {df_cleaned.shape[1]} colonnes")
print(f"\nğŸ“‹ Colonnes conservÃ©es :")
for i, col in enumerate(df_cleaned.columns, 1):
    print(f"   {i}. {col}")

# Sauvegarde
print(f"\nğŸ’¾ Sauvegarde dans : {output_file}")
df_cleaned.to_csv(output_file, sep=';', index=False, encoding='utf-8')

print(f"âœ… Fichier nettoyÃ© sauvegardÃ© !")
print(f"ğŸ“Š Taille du nouveau fichier : {output_file.stat().st_size / (1024**3):.2f} GB")
print(f"ğŸ“‰ RÃ©duction de taille : {(1 - output_file.stat().st_size / input_file.stat().st_size) * 100:.1f}%")
```
